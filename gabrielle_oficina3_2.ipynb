# 2 - Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível diferenciar o país de origem (‘origin’). Sabendo disso, existe uma correlação entre a aceleração e o consumo? De que tipo? A correlação é a mesma para os diferentes países de origem? Você pode escolher a técnica que julgar mais adequada!

```python
sns.relplot(x="acceleration", y="mpg", col = 'origin', data=mpg)
```

Existe uma correlação entre a aceleração e o consumo? De que tipo? A correlação é a mesma para os diferentes países de origem?
<b>Resposta:</b> Existe sim uma correlação entre aceleração e consumo de combustíveis especialmente quando se trata de carros norte americanos. Que, desde os primeiros quilometros já começam a subir o consumo consideravelmente.

# 3 - Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível visualizar o ano do modelo (model_year). Existe alguma relação entre as três variáveis?

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre a aceleração, consumo de combustível e o ano do modelo')
sns.scatterplot(x="acceleration", y="mpg", hue = 'model_year', data=mpg)
plt.show()
```

Existe alguma relação entre as três variáveis?Existe uma tendencia de que carros mais velhos consomem mais mesmo em baixas velocidades. Contudo, os pontos convergem no meio do gráfico apontando que a relação consumo e aceleração estão bem próximos tanto em carros mais velhos como em carros mais novos.
