```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
filmes_uri = 'https://raw.githubusercontent.com/emanueles/datavis-course/master/assets/files/observable/movies.json'
filmes = pd.read_json(filmes_uri)
filmes.head()
```

<b> 2. Crie uma visualização que mostre o orçamento de todos os filmes do conjunto de dados. Qual o filme com maior orçamento e com menor orçamento?

```python
sns.set(style="darkgrid")
plt.figure(figsize=(15,8))
plt.title('Orçamento dos filmes')
sns.barplot(x="Budget_M", y="Film", color = 'orange', estimator = sum, ci = None, data=filmes)
plt.show()
```

*2.1 Filme com maior orçamento: Pirates of the Caribbean: At World's End*

*2.2 Filme com menor orçamento: The Twilight Saga: New Moon*

```python
# confirmando os dados na tabela.
filmes.sort_values(by='Budget_M', ascending = False)
```

<b> 3. Crie uma visualização que mostre a bilheteria de todos os filmes do conjunto de dados. Qual o filme com maior bilheteria e com menor bilheteria?

```python
sns.set(style="darkgrid")
plt.figure(figsize=(15,8))
plt.title('Bilheteria dos filmes')
sns.barplot(x="Worldwide_Gross_M", y="Film", color = 'g', estimator = sum, ci = None, data=filmes)
plt.show()
```

*3.1 Filme com maior bilheteria: Avatar*

*3.2 Filme com menor bilheteria: Ratatouille*

```python
# confirmando os dados na tabela.
filmes.sort_values(by='Worldwide_Gross_M', ascending = False)
```

<b> 4. Crie uma visualização que mostre a quantidade de filmes por gênero em cada ano. Em que anos foram feitos menos filmes de ação? E em que ano foram feitos menos filmes de animação?

```python
plt.figure(figsize=(15,8))
plt.title('Filmes por gênero em cada ano')
sns.countplot(x="Year", hue= 'Genre', data=filmes)
plt.show()
```

*E em que ano foram feitos menos filmes de animação?*
